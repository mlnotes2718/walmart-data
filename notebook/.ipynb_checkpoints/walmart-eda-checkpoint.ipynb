```python
from sklearn.compose import ColumnTransformer
from sklearn.impute import SimpleImputer
from sklearn.pipeline import Pipeline, make_pipeline
from sklearn.preprocessing import OneHotEncoder, StandardScaler

from sklearn.model_selection import train_test_split

from sklearn.ensemble import RandomForestRegressor
from sklearn.linear_model import Lasso, Ridge

from sklearn.model_selection import cross_val_score
from sklearn.metrics import mean_squared_error, r2_score
```

```python
df.head()
```

```python
numeric_prep = Pipeline(
    steps=[('imputer', SimpleImputer(strategy='median')), 
           ('scalar', StandardScaler())]
)

cat_prep = Pipeline(
    steps=[('imputer', SimpleImputer(strategy='constant', fill_value='missing')), 
           ('onehot', OneHotEncoder(handle_unknown='ignore', drop='first'))]
)
```

```python
cat = ['Store', 'Dept', 'Type', 'IsHoliday', 'Month']
num = ['Size', 'Temperature', 'Fuel_Price', 'CPI', 'Unemployment', 'Year']

prep = ColumnTransformer(
    [
        ('categorical', cat_prep, cat),
        ('numerical', numeric_prep, num)
    ]
)
prep
```

```python
df_sample = df.sample(5000, random_state=4)

X = df_sample[cat+num]
y = df_sample['Weekly_Sales']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=40)
```

```python
y
```

```python
from sklearn.metrics import mean_squared_error, r2_score
rf = RandomForestRegressor(random_state=4)
ls = Lasso(random_state=4)
rd = Ridge(random_state=4)

models_to_try = [rf, ls, rd]
for model in models_to_try:
    pipeline = Pipeline(steps=[('preprocessor', prep), ('model', model)])
    pipeline.fit(X_train, y_train)
    y_pred = pipeline.predict(X_test)
    
    mse = mean_squared_error(y_test, y_pred)
    r2 = r2_score(y_test, y_pred)
    
    print(f"Model: {model.__class__.__name__}")
    print(f"MSE: {mse:.2f}")
    print(f"R2 Score: {r2:.2f}")
    print()
```

```python
import warnings
warnings.filterwarnings('ignore')

all_pipelines = {}

for m in models_to_try:
    all_pipelines[m] = pipeline = make_pipeline(prep, m)
    cross_val_scores = cross_val_score(pipeline, X_train, y_train, cv=4, scoring='r2')
    print(m)
    print(f"Cross Validation Scores: {cross_val_scores}")
    print(f"Mean Cross Validation Scores: {np.mean(cross_val_scores)}")
```

```python
all_pipelines[rf]
```

```python
all_pipelines[rf].fit(X_train, y_train)
y_pred = all_pipelines[rf].predict(X_test)
rmse = np.sqrt(mean_squared_error(y_pred, y_test))
r2 = r2_score(y_pred, y_test)
print(f'Test RMSE: {rmse}')
print(f'Test R2: {r2}')
```
